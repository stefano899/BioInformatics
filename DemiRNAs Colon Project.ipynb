```python
import pandas as pd
print(pd.__version__)
```

# First File

```python
data = pd.read_excel('Results_Export_Gruppi biologici_FIXED.xlsx')
data
```

### Results: Up-Regulated miRNAs = 24

```python
# Condition for UP-regulated miRNAs: RQ > 2 AND p-value <= 0.05
up_condition = (data['RQ'] > 2) & (data['P-Value'] <= 0.05)
up_mirnas = data[up_condition].copy()
num_up_regulated = len(up_mirnas)
print(f"Number of Different Up-regulated miRNAs: {num_up_regulated}")
```

```python
num_target_upmirnas = up_mirnas['Target Name'].nunique()
target_upmirnas = up_mirnas['Target Name'].unique()
print(f"Unique Up-regulated miRNAs: {target_upmirnas}")
print(f"Number of Unique Up-regulated miRNAs: {num_target_upmirnas}")
```

### Results: Down-Regulated miRNAs = 14

```python
# Condition for DOWN-regulated miRNAs: RQ < 0.5 AND p-value <= 0.05
down_condition = (data['RQ'] < 0.5) & (data['P-Value'] <= 0.05)
down_mirnas = data[down_condition].copy()
num_down_regulated = len(down_mirnas)
print(f"Number of Different Down-regulated miRNAs: {num_down_regulated}")
```

```python
num_target_down_mirnas = down_mirnas['Target Name'].nunique()
target_down_mirnas = down_mirnas['Target Name'].unique()
print(f"Unique Down-regulated miRNAs: {target_down_mirnas}")
print(f"Number of Unique Down-regulated miRNAs: {num_target_down_mirnas}")
```
