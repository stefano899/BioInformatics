```python
import pandas as pd

tar = pd.read_csv("Homo_sapiens_TarBase-v9.tsv.gz", sep="\t", compression="gzip", low_memory=False)
```

```python
tar.mirna_name
```

```python
import re

def pulisci_mirna(nome):
    # Rimuove l'ID numerico alla fine (es. -000435, -002198, ecc.)
    return re.sub(r'-\d{6}$', '', nome)


up_mirnas = [  pulisci_mirna(seq) for seq in df_up["Target Name"].unique()]
down_mirnas = [  pulisci_mirna(seq) for seq in df_down["Target Name"].unique()]
```

```python
print(f"UP: {up_mirnas}, \nDOWN: {down_mirnas}")
```

```python
target_up = tar[tar["mirna_name"].isin(up_mirnas)][["mirna_name", "gene_name"]].dropna(subset=["gene_name"])
target_down = tar[tar["mirna_name"].isin(down_mirnas)][["mirna_name", "gene_name"]].dropna(subset=["gene_name"])
```

```python
SonicHedgehog_pathway_genes = ["HDAC1",   "CUL3"     ,"KCASH1" ,
"KCASH2",  "KCASH3"   , "GLI1" , "GLI2" ,   "GLI3",     "PTCH1", 
"PTCH2" ,"SMO" , "SHH" ,"IHH", "DHH" ,"SUFU", "HHIP"]
```

```python
target_up_Hh = target_up[ target_up["gene_name"].isin(SonicHedgehog_pathway_genes) ]
target_down_Hh = target_down[ target_down["gene_name"].isin(SonicHedgehog_pathway_genes) ]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Costruisci il grafo (es. solo up per semplicità)
G = nx.DiGraph()
for _, row in target_up_Hh.iterrows():
    G.add_edge(row["mirna_name"], row["gene_name"])

# Layout e colori
pos = nx.spring_layout(G, seed=42)
node_colors = ["#ff9999" if n.startswith("hsa-") else "#99ccff" for n in G.nodes()]

plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color=node_colors, font_size=9, arrows=True)
plt.title("miRNA → Target mRNA Network (Up-regolati)", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Costruisci il grafo (es. solo up per semplicità)
G = nx.DiGraph()
for _, row in target_down_Hh.iterrows():
    G.add_edge(row["mirna_name"], row["gene_name"])

# Layout e colori
pos = nx.spring_layout(G, seed=42)
node_colors = ["#ff9999" if n.startswith("hsa-") else "#99ccff" for n in G.nodes()]

plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color=node_colors, font_size=9, arrows=True)
plt.title("miRNA → Target mRNA Network (Down-regolati)", fontsize=16)
plt.tight_layout()
plt.show()
```
