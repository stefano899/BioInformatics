# ANALISI GENES-MIRNA UP E DOWN DATA LA LISTA DI GENI

```python
import re

def pulisci_mirna(nome):
    # Rimuove l'ID numerico alla fine (es. -000435, -002198, ecc.)
    return re.sub(r'-\d{6}$', '', nome)


up_mirnas = [  pulisci_mirna(seq) for seq in df_up["Target Name"].unique()]
down_mirnas = [  pulisci_mirna(seq) for seq in df_down["Target Name"].unique()]
```

# PLOT ANALISI GENI-MIRNA

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite

# Filtra le interazioni up‑regolate
gene_set = set(gene_list)
gene_up = tar[
    (tar['gene_name'].isin(gene_set)) &
    (tar['mirna_name'].isin(up_mirnas))
]

# Costruisci il grafo bipartito
G_up = nx.Graph()
for _, row in gene_up.iterrows():
    G_up.add_node(row['mirna_name'], bipartite=0)
    G_up.add_node(row['gene_name'], bipartite=1)
    G_up.add_edge(row['mirna_name'], row['gene_name'])

# Calcola il layout (miRNA a sinistra)
mirna_nodes_up = [n for n, d in G_up.nodes(data=True) if d['bipartite']==0]
pos_up = bipartite_layout(G_up, mirna_nodes_up)

# Disegna
plt.figure(figsize=(12, 8))
nx.draw(G_up, pos_up,
        with_labels=True,
        node_size=200,
        font_size=6)
plt.title("Network Bipartito: miRNA Up‑regolati (sinistra) ↔ Geni (destra)")
plt.axis('off')
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
plt.show()
```

```python
# 1) Calcola il grado di ogni nodo
degree_dict = dict(G_up.degree())

# 2) (Opzionale) Trasforma in DataFrame per una visualizzazione più comoda
import pandas as pd

deg_df = (
    pd.DataFrame.from_dict(degree_dict, orient='index', columns=['degree'])
      .reset_index()
      .rename(columns={'index': 'node'})
)

# Se vuoi distinguere miRNA vs gene:
deg_df['type'] = deg_df['node'].apply(
    lambda x: 'miRNA' if x in mirna_nodes_down else 'gene'
)

# 3) Stampa a video
print(deg_df.sort_values(['type','degree'], ascending=[True, False]))
```

```python
SonicHedgehog_pathway_genes = ["HDAC1",   "CUL3"     ,"KCASH1" ,
"KCASH2",  "KCASH3"   , "GLI1" , "GLI2" ,   "GLI3",     "PTCH1", 
"PTCH2" ,"SMO" , "SHH" ,"IHH", "DHH" ,"SUFU", "HHIP"]
```

```python
target_up_Hh = target_up[ target_up["gene_name"].isin(SonicHedgehog_pathway_genes) ]
target_down_Hh = target_down[ target_down["gene_name"].isin(SonicHedgehog_pathway_genes) ]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Costruisci il grafo (es. solo up per semplicità)
G = nx.DiGraph()
for _, row in target_up_Hh.iterrows():
    G.add_edge(row["mirna_name"], row["gene_name"])

# Layout e colori
pos = nx.spring_layout(G, seed=42)
node_colors = ["#ff9999" if n.startswith("hsa-") else "#99ccff" for n in G.nodes()]

plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color=node_colors, font_size=9, arrows=True)
plt.title("miRNA → Target mRNA Network (Up-regolati)", fontsize=16)
plt.tight_layout()
plt.show()
```
